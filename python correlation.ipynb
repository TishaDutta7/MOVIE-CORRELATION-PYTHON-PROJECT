```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your original DataFrame is named 'df'

# 1. Select only numeric columns: This resolves the initial ValueError.
numeric_df = df.select_dtypes(include=['number'])

# 2. Calculate the correlation matrix on the numeric data
correlation_matrix = numeric_df.corr()

# 3. Create the heatmap for visualization using the 'magma' colormap
# 'magma' provides a dark, sequential color scheme similar to your first image.
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    cmap='magma',  # <--- Changed colormap here
    fmt=".2f"
) 

# 4. Set titles and labels
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")

# 5. Display the plot
plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
